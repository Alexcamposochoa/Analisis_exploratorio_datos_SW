```python
mapeo_franquicias = {
    101: "Star Wars",
    102: "Jurassic Park",
    103: "Harry Potter",
    104: "El seÃ±or de los anillos",
    105: "U.C Marvel",
    106: "Indiana Jones"}
```

```python
df["Franquicias"] = df["FranchiseID"].map(mapeo_franquicias)
```

```python
df = df[["MovieID","Title","Studio","Franquicias","Year","ReleaseDate","Runtime","Lifetime Gross","Budget","puntuacion","cantidad de votos","FranchiseID",]]
```

```python
franquicias = df.groupby('Franquicias')
```

```python
df_franquicias = franquicias.agg({"Lifetime Gross": ["sum", "mean"],
                                    "Budget": ["sum", "mean"],
                                    "Rentabilidad": "sum",
                                    "Indice de rentabilidad": "mean",
                                    "puntuacion": "mean"}).reset_index()
```

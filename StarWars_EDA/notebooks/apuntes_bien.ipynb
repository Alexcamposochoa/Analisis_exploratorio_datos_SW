```python
mapeo_franquicias = {
    101: "Star Wars",
    102: "Jurassic Park",
    103: "Harry Potter",
    104: "El seor de los anillos",
    105: "U.C Marvel",
    106: "Indiana Jones"}
```

```python
df_merged["Franquicias"] = df_merged["FranchiseID"].map(mapeo_franquicias)
```

```python
df_merged.head()
```

```python
df_merged = df_merged[["MovieID","Title","Studio","Franquicias","Year","ReleaseDate","Runtime","Lifetime Gross","Budget","Profitability","Profitability index '%'","puntuacion","cantidad de votos","FranchiseID",]]
```

```python
df_merged.head(10)
```

```python
df = df_merged
```

```python
df.groupby(["Franquicias","Title"])["Lifetime Gross"].aggregate(["mean","median","max"])
```

```python
franquicias = df.groupby('Franquicias')
```

```python
df_franquicias = franquicias.agg({
    "Lifetime Gross": ["sum", "mean"],
    "Budget": ["sum", "mean"],
    "Profitability": "sum",
    "Profitability index '%'": "mean",
    "puntuacion": "mean"}).reset_index()
```

```python
df_franquicias.columns = ['Franquicias', 'Suma Facturacin', 'Media Facturacin', 'Suma Presupuesto', 'Media Presupuesto', 'Rentabilidad Total', 'ndice de Rentabilidad', 'Puntuacin Media']
```

```python
numero_peliculas = [6,8,5,6,11,29]
```
